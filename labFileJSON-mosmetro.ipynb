```python
import pandas as pd
```

```python
data = pd.read_json('mosmetro.json', orient='records')
```

- вывести в консоль список статусов станций (т.е. уникальных)

```python
print(data['ObjectStatus'].unique())
```

- вывести в консоль список статусов станций и количеством таких станций

```python
print(data.groupby('ObjectStatus').count()['Station'].to_string())
```

- вывести в консоль список станций в алфавитном порядке, где статус станции отличается от 'действует'


```python
print(data.loc[data['ObjectStatus'] != 'действует'].sort_values('Station')['Station'].to_string())
```

- вывести в консоль список станций в алфавитном порядке, которые имеют выход к железнодорожному вокзалу (см. RailwayTerminal)


```python
print(data.loc[data['RailwayStation'].str.len() != 0].sort_values('Station')['Station'].to_string())
```

- вывести в консоль список железнодорожных вокзалов в алфавитном порядке (см. RailwayTerminal)


```python
print(data.loc[data['RailwayStation'].str.len() != 0].sort_values('RailwayStation', key=lambda x: x.str['value'])['RailwayStation'].to_string())
```

- вывести в консоль список районов, в которых метрополитен имеет выходы к станциям МЦД (см. MCDStation)


```python
print(data.loc[data['MCDStation'].str.len() != 0]['District'].to_string())
```

- вывести в консоль список веток метрополитена и количеством станций на каждой ветке

```python
print(data.groupby('Line').count()['Station'].to_string())
```

- вывести в консоль список административных округов в алфавитном порядке и количеством станций

```python
print(data.sort_values('AdmArea').groupby('AdmArea').count()['Station'].to_string())
```

- вывести в консоль список станций в порядке убывания количества символов в названии станции

```python
print(data.sort_values('Station', key=lambda x: x.str.len(), ascending=False)['Station'].to_string())
```

- вывести в консоль список количество символов по возрастанию и количеством станций указанной длины

```python
for length in range(data['Station'].str.len().max()):
    print(length, data.loc[data['Station'].str.len() == length]['Station'].to_string())
    print()
```

- вывести в консоль список букв русского алфавита в алфавитном порядке и количеством станций, начинающихся на каждую букву


```python
alphabet = 'йцукенгшщзхъфывапролджэячсмитьбю'.upper()
for char in sorted(alphabet):
    print(char, data[data['Station'].str.startswith(char)]['Station'].count())
```

- создать json-файл, где будут храниться все станции метро, в названии которых одно слово (т.е. нет пробелов)


```python
data.loc[~data['Station'].str.contains(' ')].to_json('stations_without_spaces.json', orient='records', force_ascii=False)
```

- создать json-файл, где будут храниться только названия станций, линий и статусы станций


```python
data[['Station', 'Line', 'ObjectStatus']].to_json('stations_lines_statuses.json', force_ascii=False, orient='records')
```

- создать N json-файлов с данными по каждой ветке метро (например, "temp\Калининская линия.json"), в котором будет отсутствовать столбец ветка ('Line')


```python
for i, row in data.groupby('Line'):
    line = row['Line'].values[0]
    row.drop(['Line'], axis=1, inplace=True)
    row.to_json(f'temp/{line}.json', force_ascii=False, orient='records')
```

- создать json-файл, где в виде двухуровневого списка хранить данные, основной список это ветки метро, а у каждой ветки будет вложенный список это станции с остальными параметрами, кроме названия ветки метро


```python
data.groupby('Line').apply(lambda x:x[["ID","Station","AdmArea","District","MCDStation","AeroexpressStation","RailwayStation","ObjectStatus","global_id"]].to_dict()).reset_index(name='Lines').to_json('lines.json', force_ascii=False, orient='records')
```
