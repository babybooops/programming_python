```python
for i, row in data.groupby('Регион'):
    region = row['Регион'].values[0]
    row.drop(['Регион'], axis=1, inplace=True)
    row.to_csv(f'temp/cities-{region}.csv')
```

```python
alphabet = 'йцукенгшщзхъфывапролджэячсмитьбю'.upper()
for char in sorted(alphabet, reverse=True):
    print(char, data[data['Город'].str.startswith(char)]['Город'].count())
```

```python
for length in range(data['Город'].str.len().max()):
    print(length, data.loc[data['Город'].str.len() == length]['Город'])
```

```python
import re
max_vowels = {}
for i, row in data.iterrows():
    max_vowels[row['Город']] = len(re.findall(r'[уеыаоэяию]{1}', row['Город'].lower()))
print(sorted(max_vowels.items(), key=lambda x:x[1], reverse=True)[:10])
```

```python
alphas = {}
for i, row in data.iterrows():
    for alpha in row['Город']:
        if alpha not in alphas.keys():
            alphas[alpha] = 1
        else:
            alphas[alpha] += 1
print(sorted(alphas.items(), key=lambda x:x[1], reverse=True)[:10])
```
