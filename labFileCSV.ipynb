```python
import csv
```

```python
# в кодировке Windows-1251
file_cities_csv = "cities.csv"
```

```python
with open(file_cities_csv) as f:
    reader = csv.reader(f)
    for line in reader:
        print(line)
```

```python
with open(file_cities_csv) as f:
    reader = csv.DictReader(f, delimiter=';')
    for line in reader:
        print(line['Город'])
```

```python
with open(file_cities_csv) as f:
    reader = csv.DictReader(f, delimiter=';')
    for line in reader:
        if "Брянская область" in line['Регион']:
            print(line['Город'])
```

```python
temp_header= []
temp_data = []
with open(file_cities_csv) as f:
    reader = csv.DictReader(f, delimiter=';')
    temp_header = reader.fieldnames
    for line in reader:
        if "Брянская область" in line['Регион']:
            temp_data.append(line)

print(temp_header)
for line in temp_data:
    print(line.values())
```

```python
from tempdir import tempdir
```

```python
with open(tempdir("temp_cities_bryansk"), 'w') as f:
    writer = csv.DictWriter(f, lineterminator='\n', delimiter=';', fieldnames = temp_header)
    writer.writeheader()
    writer.writerows(temp_data)
```

```python
with open(file_cities_csv) as f:
    reader = csv.DictReader(f, delimiter=';')
    for line in reader:
        if '-' in line ['Город']:
            print(line['Город'])
```

```python
temp_district = []

with open(file_cities_csv) as f:
    reader = csv.DictReader(f, delimiter=';')
    for line in reader:
        temp_district.append(line['Федеральный округ'])

print(*set(temp_district), sep='\n')
```
