обрезка изображений через Pillow (crop)

```python
image = Image.open('sova.jpg')
cropped = image.crop((0, 80, 200, 400))
cropped.save('temp/cropped_jelly.png')
```

использование фильтров в Pillow

```python
from PIL import ImageFilter
from PIL import Image

image = Image.open('sova.jpg')
blurred_jelly = image.filter(ImageFilter.BLUR)
blurred_jelly.save('temp/blurry_sova.png')
```

рисуем фигуры в Pillow: ellipse, rectangle и line

```python
from PIL import Image, ImageDraw
 
# Пустой желтый фон.
im = Image.new('RGB', (500, 300), (219, 193, 27))
draw = ImageDraw.Draw(im)
 
# Рисуем красный эллипс с черной оконтовкой.
draw.ellipse((100, 100, 150, 200), fill='red', outline=(0, 0, 0))
 
# Рисуем синий прямоугольник с белой оконтовкой.
draw.rectangle((200, 100, 300, 200), fill='blue', outline=(255, 255, 255))
 
# Рисуем розовую линию с шириной в 10 пиксель.
draw.line((350, 200, 450, 100), fill='pink', width=10)
 
im.save('draw-ellipse-rectangle-line.jpg', quality=95)
```

работа со шрифтами

```python
from PIL import Image, ImageDraw, ImageFont

out = Image.new("RGB", (500, 250), (255, 255, 255))
fnt = ImageFont.truetype("/usr/share/fonts/truetype/freefont/FreeSans.ttf", size=50)
d = ImageDraw.Draw(out)

# многострочный текст
d.multiline_text((100, 100), "Hello\nWorld", font=fnt, fill=(50, 50, 50))
out.save('multiline_text.png')
```
